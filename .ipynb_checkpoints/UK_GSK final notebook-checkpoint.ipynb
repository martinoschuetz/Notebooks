# 3. Unsupervised learning

### Generate a CAS table with a ID variable which is ordered by date and seq_no

```python
ordered = sess.CASTable("labelledHubAngle",caslib="public")
ordered = ordered[["newID","date_img_taken","seq_no"]]
ordered2 = ordered.sort_values(by=["date_img_taken","seq_no","newID"],ascending=True)
sess.partition(casout=dict(name="hubAngleSorted",replace=True), table=ordered2)
sess.datastep.runcode(code="data hubanglesorted2; set hubanglesorted; where substr(date_img_taken,1,8) = '20170901'; by date_img_taken seq_no newID; retain idSort; if _n_=1 then idSort=1; else idSort=idSort+1; run;")
sess.deepLearn.dlJoin(annotatedTable={"caslib":"casuser","name":"hubanglesorted2","vars":[{"name":"newID"},{"name":"idSort"}]},
                   casOut={"caslib":"casuser","name":"hubanglesorted3","replace":True},
                   idVar="newID",
                   table={"caslib":"public","name":"labelledHubAngle"})

nRows = sess.table.recordCount(table={"name":"hubanglesorted3"})
numRows = nRows["RecordCount"].values[0,0]
```

### 3.1.1. Pairwise comparison of each image to it's predecessor

```python
sess.dropTable("resultsForOne",caslib="public")
sess.dropTable("resultsForOne",caslib="casuser")

stepSize = 1

for i in range(stepSize,numRows):
    where_s = 'idSort = '+str(i+1)
    where_r = 'idSort >= '+str(i-stepSize+1)+' and idSort <= '+str(i)
    sess.image.compareimages(casout={'name':'HubAngleCompareOverTime','replace':True},
                      pairall=True,
                      method="SSIM",
                      maximum=1,
                      minimum=0,
                      referenceimages={'table':{'name':'hubanglesorted3',"where":where_r}},
                      sourceimages={'table':{'name':'hubanglesorted3',"where":where_s}})
    if i == stepSize:
        sess.datastep.runcode(code='data resultsForOne; set HubAngleCompareOverTime; run;')
    else:
        sess.datastep.runcode(code='data resultsForOne; set resultsForOne HubAngleCompareOverTime; run;')
sess.partition(table={"name":"resultsForOne"},casout={"name":"resultsForOne",'caslib':'public',"promote":True})
```

```python
resultsForOne = sess.CASTable("resultsForOne",caslib="public")
resultsForOne["ID"] = resultsForOne['_source_id_'].str.slice(5,14)
sess.partition(table=resultsForOne,casout={"name":"resultsForOne_withID",'caslib':'casuser',"replace":True})
#sess.dropTable("resultsForOne_withLabels",caslib="public")
sess.deepLearn.dlJoin(annotatedTable={"name":"resultsForOne_withID","vars":[{"name":"ID"},{"name":"_channel1_"}]},
                   casOut={"caslib":"public","name":"resultsForOne_withLabels","promote":True},
                   idVar="ID",
                   table={"caslib":"public","name":"labelledHubAngle"})
resultsForOne_withLabels = sess.CASTable("resultsForOne_withLabels",caslib="public")
resultsForOne_withLabels.head()
```

### 4.2. RPCA

### RPCA decomposes a matrix M into a low rank matrix L and a sparse matrix S such that M = L + S. Then dimension reduction is done on L, and sparse is supposed to have anomalies and noise.

##### Flatten the image out first

```python
#sess.dropTable("flattenedImagesTable",caslib="public")
sess.image.flattenImageTable(
    casout={'name':'flattenedImagesTable',"caslib":"public","promote":True},
    table={'name':'labelledHubAngle',"caslib":"public"},
    groupChannels = True,
    w = 1624,
    h = 946)
```

##### Run data pre-processing

```python
sess.dataStep.runCode(code="""
data flattenedImagesTable_wide; 
set public.flattenedImagesTable; 
_path_ = "col"||substr(_path_,54,9); 
keep _path_ c1-c1536304; 
run;
""")
```

##### Inverse the shape of the table through transpose

```python
sess.transpose(casout={"name":'flattenedImagesTable_narrow',"replace":True},table='flattenedImagesTable_wide',id="_path_")
```

##### Run PCA analysis

```python
sess.robustpca(
      table={"name":"flattenedImagesTable_narrow"},
      method="ALM",
      decomp="svd",
      id="_name_",
      outmat={"lowrankmat":{"name":"lowrankmat", "replace":True}, 
              "sparsemat":{"name":"sparsemat", "replace":True}},
      outsvd={"svdleft":{"name":"svdleft", "replace":True}, 
              "svddiag":{"name":"svddiag", "replace":True}, 
              "svdright":{"name":"svdright", "replace":True}}
   )
```

##### Convert low rank and sparse files back into images

```python
sess.transpose(casout={"name":'lowrankmat_id',"replace":True},table='lowrankmat',id="_name_")
sess.dataStep.runCode(code="data LM_res; set lowrankmat_id; name_mod=cats(_name_,'.jpg'); length 'nm'n varchar(25);format 'nm'n $CHAR25.;'nm'n= PUT('name_mod'n,$CHAR25.);run ;")
sess.image.condenseimages(casout={"name":'LM_images',"replace":True},w=1624,h=946,copyvars={"nm"},numberOfChannels='GRAY_SCALE_IMAGE',imagetable='LM_res')
sess.image.saveImages(caslib="casuser",subDirectory='LM_result',images={"table":'LM_images',"path":"nm"},type='jpg',prefix='img')
```

```python
sess.transpose(casout={"name":'sparsemat_id',"replace":True},table='sparsemat',id="_name_") 
sess.dataStep.runCode(code="data SM_res; set sparsemat_id; name_mod=cats(_name_,'.jpg'); length 'nm'n varchar(25);format 'nm'n $CHAR25.;'nm'n= PUT('name_mod'n,$CHAR25.);run ;")
sess.image.condenseimages(casout={"name":'SM_images',"replace":True},w=1624,h=946,copyvars={"nm"},numberOfChannels='GRAY_SCALE_IMAGE',imagetable='SM_res')
sess.image.saveImages(caslib="casuser",subDirectory='SM_result',images={"table":'SM_images',"path":"nm"},type='jpg',prefix='img')
```

###### Function to crop images

```python
def getPatch(params):
    sess.image.processImages(
        casout={'name':'HA_trainPatch',"caslib":"casuser","replace":True},
        imagefunctions=params,
        table={'name':'HA_train_aug',"caslib":"public"},
        copyvars=["Good_Bad_T1_v2","_PartInd_"])
    sess.image.processImages(
        casout={'name':'HA_testPatch',"caslib":"casuser","replace":True},
        imagefunctions=params,
        table={'name':'HA_test',"caslib":"public"},
        copyvars=["Good_Bad_T1_v2"])
    table=sess.CASTable('HA_trainPatch')
    val = ImageTable.from_table(table,label_col="_id_")
    val.show(nimages=2, ncol=2, randomize=False)
    HA_trainPatch = sess.CASTable("HA_trainPatch")
    HA_trainPatch["_label_"] = HA_trainPatch["Good_Bad_T1_v2"]
    sess.partition(table=HA_trainPatch,casout={"name":"HA_trainPatch","replace":True})
    HA_testPatch = sess.CASTable("HA_testPatch")
    HA_testPatch["_label_"] = HA_testPatch["Good_Bad_T1_v2"]
    HA_testPatch["_filename_0"] = HA_testPatch["_path_"].str.slice(53,62)
    sess.partition(table=HA_testPatch,casout={"name":"HA_testPatch","replace":True})
```

# Bringing all of the scores together into a scorecard

```python
sess.fedSql.execDirect(
    query='''
        create table all_dl_approaches {options replace=true} as
        select 
            vgg16._path_
            ,vgg16._id_
            ,vgg16.Good_Bad_T1_v2
            ,vgg16._DL_PredP_ as _DL_PredP_vgg16
            ,vgg16.I_Good_Bad_T1_v2 as label_vgg16
            ,rn50c._DL_PredP_ as _DL_PredP_rn50c
            ,rn50c.I_Good_Bad_T1_v2 as label_rn50c
            ,rn50s._DL_PredP_ as _DL_PredP_rn50s
            ,rn50s.I_Good_Bad_T1_v2 as label_rn50s
            ,rn101._DL_PredP_ as _DL_PredP_rn101
            ,rn101.I_Good_Bad_T1_v2 as label_rn101
            ,rn152._DL_PredP_ as _DL_PredP_rn152
            ,rn152.I_Good_Bad_T1_v2 as label_rn152
            ,dn._DL_PredP_ as _DL_PredP_dn
            ,dn.I_Good_Bad_T1_v2 as label_dn
            ,min(vgg16._DL_PredP_,rn50c._DL_PredP_,rn50s._DL_PredP_,rn101._DL_PredP_,rn152._DL_PredP_,dn._DL_PredP_) as min_all_models
            ,case when min(vgg16._DL_PredP_,rn50c._DL_PredP_,rn50s._DL_PredP_,rn101._DL_PredP_,rn152._DL_PredP_,dn._DL_PredP_) <= 0.5 then 'defect' else 'no defects' end as I_Good_Bad_T1_v2
            ,sum(case when vgg16.I_Good_Bad_T1_v2 = 'defect' then 0 else 1 end,
                case when rn50c.I_Good_Bad_T1_v2 = 'defect' then 0 else 1 end,
                case when rn50s.I_Good_Bad_T1_v2 = 'defect' then 0 else 1 end,
                case when rn101.I_Good_Bad_T1_v2 = 'defect' then 0 else 1 end,
                case when rn152.I_Good_Bad_T1_v2 = 'defect' then 0 else 1 end,
                case when dn.I_Good_Bad_T1_v2 = 'defect' then 0 else 1 end) as sum_no_pred_no_defects
            ,sum(case when vgg16.I_Good_Bad_T1_v2 = 'defect' then 0 else 1 end,
                case when rn50c.I_Good_Bad_T1_v2 = 'defect' then 0 else 1 end,
                case when rn50s.I_Good_Bad_T1_v2 = 'defect' then 0 else 1 end,
                case when rn101.I_Good_Bad_T1_v2 = 'defect' then 0 else 1 end,
                case when rn152.I_Good_Bad_T1_v2 = 'defect' then 0 else 1 end,
                case when dn.I_Good_Bad_T1_v2 = 'defect' then 0 else 1 end)/6 as avg_no_pred_no_defects
        from Valid_Res_4w0ljp vgg16
            ,Valid_Res_KvTYVg rn50c
            ,Valid_Res_AkitC2 rn50s
            ,Valid_Res_b0LjGI rn101
            ,Valid_Res_xEkPdn rn152
            ,Valid_Res_3TWPd0 dn
        where vgg16._path_ = rn50c._path_ 
            and vgg16._path_ = rn50s._path_
            and vgg16._path_ = rn101._path_
            and vgg16._path_ = rn152._path_
            and vgg16._path_ = dn._path_
          ;''')
```
