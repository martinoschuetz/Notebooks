# Basic imports and definitions

```python
import os
import zipfile
import requests as req
import json
import random
import pprint as pp
import sys
import time

from IPython.display import display
from IPython.display import Image
from swat import *

from matplotlib import pyplot as plt
%matplotlib inline
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

root_path = os.getcwd() + '/'
print(root_path)

# Create directory for generated output
if not os.path.exists(root_path + 'output'):
    os.makedirs(root_path + 'output')

def zipdir(src, dst):
    zf = zipfile.ZipFile('%s' % (dst), "w", zipfile.ZIP_DEFLATED)
    abs_src = os.path.abspath(src)
    for dirname, subdirs, files in os.walk(src):
        for filename in files:
            absname = os.path.abspath(os.path.join(dirname, filename))
            arcname = absname[len(abs_src) + 1:]
            print ("zipping %s as %s" % (os.path.join(dirname, filename),arcname))
            zf.write(absname, arcname),
    zf.close()
```

```python
headers={
  'content-type': 'application/vnd.sas.collection+json',
  'Authorization': 'Bearer ' + myToken
}

variablesResult = req.get(mmService + variablesUri, headers = headers)
projectVariables = variablesResult.json()

# Create input and output variable collections
projectInputVars = []
projectOutputVars = []
for variable in projectVariables['items']:
    if variable['role'] == 'input':
        projectInputVars.append(variable)
    else:
        projectOutputVars.append(variable)
```

# Performance Task Definition

```python
print("Project ID = " + projectID)
print("Model ID of Champion = " + ChampionID)

# Load Performance Monitoring Data
#%run -t Lifecycle_Monitor_XGBOOST.ipynb ['--uuid', 'ChampionID']

# Script call including parameter didn't work. Thus use file for communication to other script.
fileNew=open(root_path + "output/uuid.txt","w+")
fileNew.write(ChampionID)
fileNew.close()

%run -t Lifecycle_Monitor_XGBOOST.ipynb
```

```python
# Hint: Currently the file name of the scroring Data also requires the ID of the Champion model as part of the file name.

inputVarList = []
outputVarList = []

for variable in projectInputVars:
    inputVarList.append(variable['name'])

for variable in projectOutputVars:
    outputVarList.append(variable['name'])

performanceTask = {
    'projectId': projectID,
    'modelIds': [ChampionID],
    'name' : 'PERFORMANCE_TEST',
    'inputVariables': inputVarList,
    'outputVariables': outputVarList,
    'maxBins': 10,
    "casServerId": "cas-shared-default",
    'dataLibrary' : 'Public',
    'resultLibrary' : 'ModelPerformanceData',
    'dataPrefix': 'HMEQMONITOR',
    "includeAllData": True,
    "characteristicAlert": "char_p1>2",
    "characteristicWarn": "char_p1>5 or char_p25>0",
    "stabilityAlert": "outputDeviation > 0.03",
    "stabilityWarn": "outputDeviation > 0.01",
    "assessAlert": "(lift5Decay>0.15 and lift10Decay>0.12) or giniDecay>0.1 or ksDecay>0.1",
    "accessWarn": "lift5Decay>0.05 or MCE > 0.1",
    'scoreExecutionRequired': False,
    'performanceResultSaved': True,
    'loadPerformanceResult': True,
    'forcingRunAllData': True,
    'championMonitored' : True,
    'traceOn': True
}

headers={
  'Content-type': 'application/vnd.sas.models.performance.task+json',
  'Accept': 'application/vnd.sas.models.performance.task+json',
  'Authorization': 'Bearer ' + myToken
}

url = mmService + '/modelManagement/performanceTasks?projectId=' + projectID
performanceTask = req.post(url, data=json.dumps(performanceTask), headers = headers).json()

performanceTaskId = performanceTask['id']

print("Performance task id = " + performanceTaskId)
pp.pprint(performanceTask)
```
