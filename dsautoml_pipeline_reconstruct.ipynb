```python
import swat
```

#### Start CAS session

```python
server = "rdcgrd133"
port = 16271
s = swat.CAS(server, port)
```

```python
tbl_name = 'hmeq'
target = 'bad'
tbl = s.upload('U:\\data\\hmeq.csv', casout=dict(name=tbl_name, replace=True))
```

```python
s.fetch(table=tbl_name, to=5)
```

```python
s.loadactionset(actionset='dataSciencePilot')
```

```python
s.highcardinality(table=tbl_name)
```

```python
res = s.dataSciencePilot.dsAutoml(table=tbl_name, 
                                  target=target, 
                                  # use all inputs ...or specify ...inputs=[....]
                                  inputs = ['loan', 'job', 'reason', 'clno', 'ninq'],
                                  modelTypes = ['decisionTree'],
                                  # an example of manual override of the role auto-detection
                                  # the below means use cnlno and ninq as intervals while for the rest of inputs,
                                  # do auto-detection
                                  explorationPolicy = dict(nominal=dict(intervals=['clno', 'ninq'])),
                                  sampleSize = 1,
                                  featuresOut = dict(name='fout', replace=True),
                                  transformationOut = dict(name='tout', replace=True),
                                  pipelinesOut = dict(name='pout', replace=True),
                                  saveState = dict(name='fmout', replace=True))
```

```python
s.fetch(table='fout')['Fetch']
```

```python
s.fetch(table='pout')['Fetch']
```

```python
features_df = s.fetch(table='fout')['Fetch']
pipeline_df = s.fetch(table='pout')['Fetch']
```

```python
# reconstruct the first pipeline:- basically extract the features and nominals.
def extrac_pipeline_meta_data(pipeline_df, features_df, pipeline_id):
    row = pipeline_df[pipeline_df.PipelineId == pipeline_id]
    n_feats = int(row['NFeatures'][0])
    features = []
    nominals = []
    for i in range(1, n_feats+1):
        col_name = 'Feat{}Id'.format(i)
        feat_id = row[col_name][0]
        col_name = 'Feat{}IsNom'.format(i)
        is_nom = row[col_name][0]
        feat_name = features_df[features_df.FeatureId == feat_id]['Name'][feat_id - 1]
        features.append(feat_name)
        if is_nom:
            nominals.append(feat_name)
    return features, nominals
    
```

```python
# do astore scoring of your datasets...
s.loadactionset(actionset='astore')
```

```python
featurized_tbl = 'featurized_tbl'
s.astore.score(table=tbl_name, copyvars=[target], rstore='fmout',
               casout=dict(name=featurized_tbl, replace=True))
```

```python
s.fetch(table=featurized_tbl, to=5)
```

```python
s.loadactionset(actionset='autotune')
```

```python
pipeline_id = 1
pp_features, pp_nom_features = extrac_pipeline_meta_data(pipeline_df, features_df, pipeline_id)
```

```python
pp_features
```

```python
pp_nom_features
```

```python
s.autotune.tunedecisiontree(trainoptions=dict(table=featurized_tbl,
                                              target=target,
                                              inputs=pp_features,
                                              nominals=pp_nom_features),
                            tunerOptions=dict(seed=123))
```
