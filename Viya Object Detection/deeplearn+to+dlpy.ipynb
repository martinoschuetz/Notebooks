# ResNet50
From swat deeplearn to dlpy

```python
import swat as sw
from dlpy import Model
```

```python
# Connection for ben on digits
s = sw.CAS('localhost',5570,'cas', 'Orion123')
s.loadactionset('deepLearn')
```

### deepLearn model built with swat in ResNet_deepLearn.py

```python
# swat deepLearn function for resnet50 model architecture 
import ResNet_deepLearn
```

```python
ResNet_deepLearn.ResNet50_Model_deepLearn(s, modelName='RN50_No_Freeze', nchannels=3, width=224, height=224, scale=1/255, 
                                          inputCropType="UNIQUE", inputChannelOffset=[0.406, 0.456, 0.485], randomFlip="H")
```

```python
s.tableinfo()
```

```python
model_deepLearn = s.CASTable('RN50_No_Freeze')
```

```python
model_dlpy = Model.from_table(model_deepLearn, output_model_table='model_dlpy')
```

```python
s.tableinfo()
```

```python
model_dlpy.print_summary()
```
